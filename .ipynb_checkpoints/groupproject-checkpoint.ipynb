## Stock Performance Comparison: NVDA vs. Peers
### Description: Tracks the percentage-based price movement of semiconductor stocks from mid-January to mid-March 2025.

```python
start = "2025-01-15"
end = "2025-03-15"
nvda = pd.read_csv("NVDA.csv")
nvda["ticker"] = "NVDA"
amd = pd.read_csv("AMD.csv")
amd["ticker"] = "AMD"
asml = pd.read_csv("ASML.csv")
asml["ticker"] = "ASML"
arm = pd.read_csv("ARM.csv")
arm["ticker"] = "ARM"
df = pd.concat([nvda, asml, arm, amd])
# 图一股价对比
df["norm_price"] = df["close"] / df.groupby("ticker")["close"].transform("first") * 100

df["date"] = pd.to_datetime(df["date"])
df = df[(df["date"] >= start) & (df["date"] <= end)]
fig = px.line(
    df, x="date",
    y="norm_price",
    color="ticker",
    labels={
        "date": "Date",
        "norm_price": "Normalized Price"
    },
    hover_data={
        "ticker": True,
        "norm_price": ":.2f",
        "date": True,
        "close": True,
        "open": True
    },

)
```

```python
fig.show()
```

## NVDA Correlation Analysis: Public Interest vs. Market Valuation
### Description: This scatter plot examines the relationship between NVDA's stock price and investor attention (page views), using an OLS regression line to highlight the degree of correlation.

```python
raw = requests.get(url, headers=headers)

data = raw.json()
df2 = pd.DataFrame(data["items"])
df2["date"] = pd.to_datetime(df2["timestamp"].astype(str).str[:8])
df2 = df2[["date", "views"]]
df_nvda = df[df["ticker"] == "NVDA"].copy()
df_merge = pd.merge(df_nvda, df2, on="date", how="inner")
df_merge = df_merge[(df_merge["date"] >= start) & (df_merge["date"] <= end)]
df_merge["month"] = df_merge["date"].dt.month
fig2 = px.scatter(
    df_merge,
    x="norm_price",
    y="views",
    trendline="ols",
    title="NVDA: Attention vs Stock Price",
    labels={
        "views": "Number of Views",
        "norm_price": "Normalized Price"
    },
    color = "month",
    hover_data={
        "ticker": True,
        "norm_price": True,
        "views": True,
        "date": True
    },
)
```

```python
fig2.show()
```

```python
import dash
from dash import dcc, html
from dash import Input, Output
import plotly.graph_objects as go
```

```python
app = dash.Dash(__name__)
app.layout = html.Div([
    html.H1("NVDA Dashboard", style={'textAlign': 'center'}),
    html.Label("Please Select the month:"),
    dcc.Dropdown(
        id='month-selector',
        options=[
            {'label': 'Jan', 'value': 'JAN'}, 
            {'label': 'Feb', 'value': 'FEB'},
            {'label': 'Mar', 'value': 'MAR'},
            {'label': 'All', 'value': 'ALL'}
        ],
        value='ALL',
        clearable=False,
        multi=False
    ),

    html.Div([
        dcc.Graph(id='fig1', figure=fig)
    ]),

    html.Div([
        dcc.Graph(id='fig2', figure=fig2)
    ])
])

@app.callback(
    [Output('fig1', 'figure'),
     Output('fig2', 'figure')],
    [Input('month-selector', 'value')]
)
def update_graphs(selected_period):

    ranges = {
        'ALL': ['2025-01-15', '2025-03-15'],
        'JAN': ['2025-01-15', '2025-01-31'],
        'FEB': ['2025-02-01', '2025-02-28'],
        'MAR': ['2025-03-01', '2025-03-15']
    }
    start_dt, end_dt = ranges.get(selected_period)

    filtered_df = df[(df["date"] >= start_dt) & (df["date"] <= end_dt)]
    
    filtered_df_merge = df_merge[(df_merge["date"] >= start_dt) & (df_merge["date"] <= end_dt)]

    new_fig1 = px.line(
        filtered_df, x="date", y="norm_price", color="ticker",
        title="Normalized Stock Price Performance"
    )

    new_fig2 = px.scatter(
        filtered_df_merge, x="norm_price", y="views", 
        trendline="ols", color="month",
        title="NVDA: Attention vs Stock Price"
    )

    new_fig1.update_layout(transition_duration=500)
    new_fig2.update_layout(transition_duration=500)

    return new_fig1, new_fig2
```

```python
app.run(debug=True, jupyter_mode="external")
```
